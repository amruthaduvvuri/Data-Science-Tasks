```python
#importing necessary libraries to read the dataframe
#first of pandas
import pandas as pd
```

```python
#getting already present dataset from colab with the help of sklearn.datasets
from sklearn.datasets import load_diabetes
```

```python
#importing other libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#if we load the datasets from this with load_diabets() we get key - value pairs
load_diabetes()
#our output is basically target
```

```python
 #saving this as dataframe df
 df = load_diabetes()
```

```python
dataset= pd.DataFrame(df.data)
```

```python
#giving column names from the dataset
dataset.columns = df.feature_names
```

```python
dataset.head()
```

```python
#indepedent and dependent features
X = dataset
y = df.target
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.30,random_state= 42)
#
```

```python
X_train
```

```python
y_train
```

```python
#normalizing and standardizing the dataset
from sklearn.preprocessing import StandardScaler
```

```python
#initialising standard scalar
scaler = StandardScaler()
```

```python
#fitting the data for tranformation fit_transform is used to learn the scaling or hyperparameters tuned on the training dataset
scaler.fit_transform(X_train)
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
#transform is used to on the new dataset i.e the testing dataset to apply the same transformation on the data
X_train = scaler.transform(X_train)
```

```python
#importing linear regression model
from sklearn.linear_model import LinearRegression
```

```python
#cross validation - splits the training data into multi folds of test and train to check the different combo of the accuracies
from sklearn.model_selection import cross_val_score
```

```python
#creating a regression object
regression = LinearRegression()
```

```python
#calculating with hyperparameters - which is regression , X_train , y_train data
cross_val_score(regression,X_train , y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
#mse - mean squared error
mse =cross_val_score(regression,X_train , y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
#taking mean or average
np.mean(mse)
```
